```python
num_O = 0
noOidx2orig = {}
for i in range(len(orig_polycrystal)):
    elem = orig_polycrystal[i].symbol
    if elem == 'O':
        num_O += 1
        continue
    else:
        noOidx2orig[i-num_O] = i
```
