```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def load_pce_data():
    """
    Load PyApprox PCE Sobol indices from NPZ files.
    Returns:
    --------
    dict with keys for each PCE max order (4, 6, 8):
        Each contains:
            'E_main': array of main indices for E
            'E_total': array of total indices for E
            'nu_main': array of main indices for nu
            'nu_total': array of total indices for nu
            'samples': array of sample sizes
    """
    # PCE max orders and sample sizes
    pce_orders = [4, 6, 8]
    sample_sizes = [12, 16, 20, 24, 28, 32, 50, 100, 500]

    data = {}

    for order in pce_orders:
        order_data = {
            'samples': [],
            'E_main': [],
            'E_total': [],
            'nu_main': [],
            'nu_total': []
        }

        for n in sample_sizes:
            filepath = f'./data_files/analytic_pce_convergence/pyapprox_pce_{order}_samples_{n}.npz'

            try:
                npz_data = np.load(filepath)

                # main_effects shape: (3, 3), row 0 = rho, row 1 = E, row 2 = nu
                # columns: col 0 = E_effective, col 1 = stress, col 2 = max_von_Mises_stress
                # total_effects shape: (3, 3), same layout

                # Extract for max_von_Mises_stress (column index 2)
                order_data['samples'].append(n)
                order_data['E_main'].append(npz_data['main_effects'][1, 0])
                order_data['E_total'].append(npz_data['total_effects'][1, 0])
                order_data['nu_main'].append(npz_data['main_effects'][2, 0])
                order_data['nu_total'].append(npz_data['total_effects'][2, 0])

            except FileNotFoundError:
                print(f"Warning: File not found: {filepath}")
                continue

        # Convert lists to numpy arrays
        for key in ['E_main', 'E_total', 'nu_main', 'nu_total']:
            order_data[key] = np.array(order_data[key])
        order_data['samples'] = np.array(order_data['samples'])

        data[order] = order_data

    return data
```

```python
def create_error_plot(pce_data):
    ref_E_main = 0.998435
    ref_E_total = 0.998458
    ref_nu_main = 1.541665e-3
    ref_nu_total = 1.564441e-3

    fig, ax = plt.subplots(figsize=(10, 6))

    # Use viridis colormap for different PCE orders
    pce_orders = [4, 6, 8]
    colors_E = cm.GnBu(np.linspace(0.05, 0.5, len(pce_orders)))
    colors_nu = cm.OrRd(np.linspace(0.55, 1.0, len(pce_orders)))

    for idx, order in enumerate(pce_orders):
        data = pce_data[order]
        color_E = colors_E[idx]

        # Main indices - solid line
        ax.plot(data['samples'], np.abs(data['E_main'] - ref_E_main),
                color=color_E, linestyle='-', marker='o', markersize=6,
                label=f'x1 main (PCE order {order})', linewidth=2)

        # Total indices - dashed line
        ax.plot(data['samples'], np.abs(data['E_total']-ref_E_total),
                color=color_E, linestyle='--', marker='s', markersize=6,
                label=f'x1 total (PCE order {order})', linewidth=2)

        color_nu = colors_nu[idx]

        # Main indices - solid line
        ax.plot(data['samples'], np.abs(data['nu_main']-ref_nu_main),
                color=color_nu, linestyle='-', marker='o', markersize=6,
                label=f'x2 main (PCE order {order})', linewidth=2)

        # Total indices - dashed line
        ax.plot(data['samples'], np.abs(data['nu_total']-ref_nu_total),
                color=color_nu, linestyle='--', marker='s', markersize=6,
                label=f'x2 total (PCE order {order})', linewidth=2)



    ax.set_xscale('log')
    ax.set_yscale('log')

    ax.set_xlabel('Number of Training Samples', fontsize=18)
    ax.set_ylabel('MAE', fontsize=18)
    ax.set_title('Sobol Indices Error for x1(1-x2^2)', fontsize=18)
    ax.legend(loc='center right', fontsize=12, ncol=2)
    ax.tick_params(axis='both', labelsize=14)
    ax.grid(True, alpha=0.5, which="both")
    #ax.set_ylim(-0.1, 1.1)
    #ax.ticklabel_format(useOffset=False, style='plain')
    #ax.hlines(1.0, 10, 20000, colors="black", linewidth=1, alpha=0.5)

    plt.tight_layout()
    #plt.savefig('sobol_pce_E_indices.png', dpi=300, bbox_inches='tight')
    #plt.savefig('sobol_pce_E_indices.pdf', bbox_inches='tight')
    #plt.close()
```

```python
pce_data = load_pce_data()
```

```python
create_error_plot(pce_data)
```
