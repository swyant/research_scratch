```python
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open("collected_delta_Es.json") as f:
    data = json.load(f)
```

```python
data["bulk_oint"]
```

```python
# manual outlier removal
bulk_oint_vals = [v for v in data["bulk_oint"].values() if v<-9.0]
print(len(bulk_oint_vals))
bulk_oint_vals
```

```python
mean_bulkoint_deltaE = np.mean(bulk_oint_vals)
mean_bulkoint_deltaE
```

```python
segregation_energies = np.array([v-mean_bulkoint_deltaE for v in data["gb_oint"].values()])
```

```python
segregation_energies
```

```python
def plot_diverging_histogram(data, bins=30):
      """
      Create a histogram with coolwarm diverging colormap.
      Negative values are blue, positive values are red.

      Parameters:
      -----------
      data : array-like
          Input array (~200 entries)
      bins : int
          Number of bins (default 30)
      """
      fig, ax = plt.subplots(figsize=(10, 6))

      # Create histogram data
      counts, bin_edges = np.histogram(data, bins=bins)
      bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
      bin_width = bin_edges[1] - bin_edges[0]

      # Normalize bin centers symmetrically around 0 for colormap
      max_abs = max(abs(bin_centers.min()), abs(bin_centers.max()))
      normalized = bin_centers / max_abs  # Maps to [-1, 1]
      colormap_input = (normalized + 1) / 2  # Maps to [0, 1] for coolwarm

      # Get colors from coolwarm (0=blue, 0.5=white, 1=red)
      cmap = plt.cm.coolwarm
      colors = cmap(colormap_input)

      # Create bars with individual colors
      ax.bar(bin_centers, counts, width=bin_width * 0.95,
             color=colors, edgecolor='black', linewidth=0.5)

      # Add dashed vertical line at 0
      ax.axvline(x=0, color='black', linestyle='--', linewidth=2, alpha=0.7)

      ax.set_xlabel('Segregation Energies (eV)')
      ax.set_ylabel('Count')
      ax.set_title("200 Segregation Energies from Dio's polycrystal")
      #ax.grid(axis='y', alpha=0.3)

      plt.tight_layout()
      return fig, ax
```

```python
plot_diverging_histogram(segregation_energies)
```
