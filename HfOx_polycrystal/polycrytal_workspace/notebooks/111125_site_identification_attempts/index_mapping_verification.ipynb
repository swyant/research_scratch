```python
from ase.io import read as ase_read
import numpy as np
import json
```

```python
noO_polycrystal_min = ase_read("111025_min4_no0_relaxed.xyz") # shouldn't use the minimized one
orig_dio_polycrystal = ase_read("Hf.O.polycrystal.xyz")
```

```python
with open("noOidx2orig.json", "r") as f:
    index_map = json.load(f)

index_map = {int(k): v for k,v in index_map.items()}
```

```python
index_map
```

```python
noO_positions = noO_polycrystal_min.get_positions()
orig_positions = orig_dio_polycrystal.get_positions()
```

```python
for k,v in index_map.items():
    noO_pos = noO_positions[k]
    orig_pos = orig_positions[v]

    pos_match = np.all(np.isclose(noO_pos, orig_pos, atol=1e-8))
    if not pos_match:
        print(f"warning issue with noO {k} and orig {v}: {noO_pos} vs. {orig_pos}")
```
