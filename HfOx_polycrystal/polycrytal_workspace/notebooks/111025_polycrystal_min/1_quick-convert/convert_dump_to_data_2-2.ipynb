```python
from ase.io import read as ase_read
from ase.io import write as ase_write
```

```python
input_file = "min4_111025_initial.dump"
output_file = "min4_111025_first-pass_DATA"
```

```python
atoms = ase_read(input_file, format="lammps-dump-text")
```

```python
atoms
```

```python
atoms.set_chemical_symbols(["Hf"]* len(atoms))
```

```python
atoms
```

```python
ase_write(output_file, atoms, format="lammps-data", masses=True, specorder=["Hf", "O"])
```
