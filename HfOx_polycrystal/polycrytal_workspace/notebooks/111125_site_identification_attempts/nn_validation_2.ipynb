```python
import numpy as np
from ase.io import read as ase_read
from ase.io import write as ase_write
```

```python
nnlist_data = np.load("111025_min4_neighborlist_10dot4A.npz")
nn_i, nn_j, nn_S = nnlist_data["i"] , nnlist_data["j"], nnlist_data["S"]
nn_dict = {}
for k in range(len(nn_i)):
    iidx = int(nn_i[k])
    if iidx not in nn_dict:
        nn_dict[iidx] = []
    nn_dict[iidx].append(int(nn_j[k]))
```

```python
def set_nn_species(center_atom_idx, nn_dict, atoms, alt_symbol="Np"):
    alt_tag_idxs = [center_atom_idx,]
    alt_tag_idxs.extend(nn_dict[center_atom_idx])
    symbols = atoms.get_chemical_symbols()
    for i in alt_tag_idxs:
        symbols[i] = alt_symbol
    atoms.set_chemical_symbols(symbols)
    return symbols
```

```python
input_xyz = "111025_min4_no0_relaxed.xyz"
polycry_noO_min = ase_read(input_xyz)
ref_symbols = polycry_noO_min.get_chemical_symbols().copy()
```
