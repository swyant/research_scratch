```python
import numpy as np
from ase.io import read as ase_read
from ase.io import write as ase_write
```

```python
nnlist_data = np.load("111025_min4_neighborlist_10dot4A.npz")
```

```python
nn_i, nn_j, nn_S = nnlist_data["i"] , nnlist_data["j"], nnlist_data["S"]
```

```python
nn_dict = {}
for k in range(len(nn_i)):
    iidx = int(nn_i[k])
    if iidx not in nn_dict:
        nn_dict[iidx] = []
    nn_dict[iidx].append(int(nn_j[k]))
```

```python
def set_nn_tags(center_atom_idx, nn_dict, atoms):
    alt_tag_idxs = [center_atom_idx,]
    alt_tag_idxs.extend(nn_dict[center_atom_idx])
    tags = np.ones(len(atoms),dtype=int)
    tags[alt_tag_idxs] = 2
    atoms.set_tags(tags)
    return tags
```
